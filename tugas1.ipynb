Import Library CV2, Numpy, Math Plot Lib

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

Variable untuk membaca gambar

```python
eskrim = cv2.imread("/content/drive/MyDrive/Google Collab/Images/eskrim.jpg")
casper = cv2.imread("/content/drive/MyDrive/Google Collab/Images/casper.jpg")
```

Menampilkan gambar

```python
print("eskrim")
cv2_imshow(eskrim)
print("casper")
cv2_imshow(casper)
```

Menampilkan dimensi gambar

```python
print("eskrim")
print(eskrim.shape)
print("casper")
print(casper.shape)
```

Konversi warna

```python
eskrimrgb = cv2.cvtColor(eskrim, cv2.COLOR_BGR2RGB)
casperhsv = cv2.cvtColor(casper, cv2.COLOR_BGR2HSV)
```

Menampilkan gambar hasil konversi warna

```python
print("eskrim rgb")
print(eskrimrgb.shape)
cv2_imshow(eskrimrgb)
print("casper hsv")
print(casperhsv.shape)
cv2_imshow(casperhsv)
```
